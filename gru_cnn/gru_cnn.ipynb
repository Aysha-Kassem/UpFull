```python
# ============================================================
# 1️⃣ Install & Import Packages
# ============================================================
!pip install -q pandas numpy scikit-learn tensorflow matplotlib seaborn joblib

import os
import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve, precision_recall_fscore_support

import tensorflow as tf
from tensorflow.keras import layers, models, regularizers, Input, callbacks

# ============================================================
# 2️⃣ Load Dataset
# ============================================================
filename = "/content/DataSet.csv"

if not os.path.exists(filename):
    raise FileNotFoundError(f"ERROR: '{filename}' not found! Please upload it manually.")

# ============================================================
# 3️⃣ Data Preprocessing
# ============================================================
def prepare_fall_detection_data(filepath=filename, time_steps=50, test_size=0.2, random_state=42):
    try:
        df = pd.read_csv(filepath, header=[0,1])
        df.columns = ['_'.join(col).strip() for col in df.columns.values]
        current_cols = df.columns.tolist()

        # Rename accelerometer and gyro columns
        if 'WristAccelerometer_x-axis (g)' in current_cols:
            idx = current_cols.index('WristAccelerometer_x-axis (g)')
            df.rename(columns={
                current_cols[idx+1]: 'WristAccelerometer_y-axis (g)',
                current_cols[idx+2]: 'WristAccelerometer_z-axis (g)'
            }, inplace=True)

        if 'WristAngularVelocity_x-axis (deg/s)' in current_cols:
            idx = current_cols.index('WristAngularVelocity_x-axis (deg/s)')
            df.rename(columns={
                current_cols[idx+1]: 'WristAngularVelocity_y-axis (deg/s)',
                current_cols[idx+2]: 'WristAngularVelocity_z-axis (deg/s)'
            }, inplace=True)

        # Activity column
        activity_col = [col for col in df.columns if col.lower().startswith('activity')][0]

        FEATURES = [
            'WristAccelerometer_x-axis (g)', 'WristAccelerometer_y-axis (g)', 'WristAccelerometer_z-axis (g)',
            'WristAngularVelocity_x-axis (deg/s)', 'WristAngularVelocity_y-axis (deg/s)', 'WristAngularVelocity_z-axis (deg/s)'
        ]

        # Magnitude features
        df['Acc_mag'] = np.sqrt(df['WristAccelerometer_x-axis (g)']**2 +
                                df['WristAccelerometer_y-axis (g)']**2 +
                                df['WristAccelerometer_z-axis (g)']**2)
        df['Gyro_mag'] = np.sqrt(df['WristAngularVelocity_x-axis (deg/s)']**2 +
                                 df['WristAngularVelocity_y-axis (deg/s)']**2 +
                                 df['WristAngularVelocity_z-axis (deg/s)']**2)
        FEATURES += ['Acc_mag','Gyro_mag']

        # Target
        FALL_CODES = [7,8,9,10,11]
        df['Target'] = df[activity_col].apply(lambda x: 1 if x in FALL_CODES else 0)

        # Train/Test split
        X = df[FEATURES].values
        y = df['Target'].values
        X_train_raw, X_test_raw, y_train_raw, y_test_raw = train_test_split(
            X, y, test_size=test_size, random_state=random_state, stratify=y
        )

        # Scaling
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train_raw)
        X_test_scaled = scaler.transform(X_test_raw)
        joblib.dump(scaler, "scaler.save")
        print("✅ Scaler saved")

        # Create sequences
        def create_sequences(data, target, steps, step_size=1):
            X_seq, y_seq = [], []
            for i in range(0, len(data)-steps, step_size):
                X_seq.append(data[i:i+steps])
                y_seq.append(target[i+steps-1])
            return np.array(X_seq), np.array(y_seq)

        X_train, y_train = create_sequences(X_train_scaled, y_train_raw, time_steps)
        X_test, y_test = create_sequences(X_test_scaled, y_test_raw, time_steps)

        print("✅ Preprocessing finished")
        print("Train shape:", X_train.shape, "Test shape:", X_test.shape)

        return X_train, X_test, y_train, y_test

    except Exception as e:
        print(f"❌ Error preprocessing: {e}")
        return None, None, None, None

X_train, X_test, y_train, y_test = prepare_fall_detection_data(filename)
if X_train is None:
    raise Exception("Data preprocessing failed!")

# ============================================================
# 4️⃣ Compute Class Weights
# ============================================================
neg_count = np.sum(y_train == 0)
pos_count = np.sum(y_train == 1)
total = len(y_train)
class_weight = {0: total/(2*neg_count), 1: total/(2*pos_count)}
print(f"Class weights: {class_weight}")

# ============================================================
# 5️⃣ Build Optimized GRU + CNN Model
# ============================================================
def build_gru_cnn_model(input_shape,
                        conv_filters=[64,128],
                        gru_units=[128,64],
                        dropout_rate=0.4,
                        l2_reg=1e-3):
    inp = Input(shape=input_shape)

    # Conv1D + BatchNorm + MaxPooling
    x = layers.Conv1D(conv_filters[0], kernel_size=3, padding='same', activation='relu',
                      kernel_regularizer=regularizers.l2(l2_reg))(inp)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling1D(2)(x)

    x = layers.Conv1D(conv_filters[1], kernel_size=3, padding='same', activation='relu',
                      kernel_regularizer=regularizers.l2(l2_reg))(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling1D(2)(x)

    # Bidirectional GRU + BatchNorm + Dropout
    x = layers.Bidirectional(layers.GRU(gru_units[0], return_sequences=True,
                                        kernel_regularizer=regularizers.l2(l2_reg)))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(dropout_rate)(x)

    x = layers.Bidirectional(layers.GRU(gru_units[1], return_sequences=False,
                                        kernel_regularizer=regularizers.l2(l2_reg)))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(dropout_rate)(x)

    # Dense layers
    x = layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(l2_reg))(x)
    x = layers.Dropout(dropout_rate/2)(x)

    out = layers.Dense(1, activation='sigmoid')(x)

    model = models.Model(inputs=inp, outputs=out, name="Fall_Detection_GRU_CNN")

    model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy',
                           tf.keras.metrics.Precision(name='precision'),
                           tf.keras.metrics.Recall(name='recall')])

    return model

input_shape = (X_train.shape[1], X_train.shape[2])
model = build_gru_cnn_model(input_shape)
model.summary()

# ============================================================
# 6️⃣ Callbacks
# ============================================================
early_stop = callbacks.EarlyStopping(monitor='val_loss', patience=7, restore_best_weights=True)
reduce_lr = callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6)

# ============================================================
# 7️⃣ Train Model
# ============================================================
history = model.fit(X_train, y_train,
                    validation_data=(X_test, y_test),
                    epochs=50,
                    batch_size=64,
                    class_weight=class_weight,
                    callbacks=[early_stop, reduce_lr],
                    verbose=1)

# ============================================================
# 8️⃣ Evaluate Model
# ============================================================
y_pred_prob = model.predict(X_test)
y_pred_default = (y_pred_prob > 0.5).astype(int).flatten()

print("\nClassification Report (threshold=0.5):")
print(classification_report(y_test, y_pred_default, target_names=['Non-Fall', 'Fall'], digits=4))

roc_auc = roc_auc_score(y_test, y_pred_prob)
print(f"ROC-AUC Score: {roc_auc:.4f}")

cm_default = confusion_matrix(y_test, y_pred_default)
print("\nConfusion Matrix (threshold=0.5):")
print(cm_default)

# ============================================================
# 9️⃣ Find Best Threshold for Fall F1
# ============================================================
best_thr = 0.5
best_f1 = -1
thresholds = np.linspace(0.1, 0.9, 81)

for thr in thresholds:
    y_pred_thr = (y_pred_prob >= thr).astype(int).flatten()
    _, _, f1, _ = precision_recall_fscore_support(y_test, y_pred_thr, average=None, labels=[0,1])
    fall_f1 = f1[1]
    if fall_f1 > best_f1:
        best_f1 = fall_f1
        best_thr = thr

y_pred_best = (y_pred_prob >= best_thr).astype(int).flatten()
cm_best = confusion_matrix(y_test, y_pred_best)

print(f"\nBest threshold for Fall F1: {best_thr:.2f}")
print("Confusion Matrix at best threshold:")
print(cm_best)

# ============================================================
# 🔟 Visualizations
# ============================================================
fig = plt.figure(figsize=(16,10))

# Confusion Matrix default
plt.subplot(2,3,1)
sns.heatmap(cm_default, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Non-Fall','Fall'], yticklabels=['Non-Fall','Fall'])
plt.title('Confusion Matrix (thr=0.5)')

# Confusion Matrix best threshold
plt.subplot(2,3,2)
sns.heatmap(cm_best, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Non-Fall','Fall'], yticklabels=['Non-Fall','Fall'])
plt.title(f'Confusion Matrix (best thr={best_thr:.2f})')

# ROC Curve
plt.subplot(2,3,3)
fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
plt.plot(fpr, tpr, label=f'AUC={roc_auc:.4f}')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('FPR'); plt.ylabel('TPR'); plt.title('ROC Curve'); plt.legend()

# Loss
plt.subplot(2,3,4)
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Val')
plt.title('Loss'); plt.legend()

# Accuracy
plt.subplot(2,3,5)
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.title('Accuracy'); plt.legend()

# Precision & Recall
plt.subplot(2,3,6)
plt.plot(history.history['precision'], label='Train Prec')
plt.plot(history.history['val_precision'], label='Val Prec')
plt.plot(history.history['recall'], label='Train Rec')
plt.plot(history.history['val_recall'], label='Val Rec')
plt.title('Precision & Recall'); plt.legend()

plt.tight_layout()
plt.show()

# ============================================================
# 1️⃣1️⃣ Save Model
# ============================================================
model.save('fall_detection_gru_cnn_optimized.h5')
model.save('fall_detection_gru_cnn_optimized.keras')

print("✓ Model saved successfully!")
```

```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import load_model
from sklearn.preprocessing import RobustScaler
import joblib

MODEL = load_model("fall_cnn_gru_model.h5")
scaler = joblib.load("scaler.save")

print("Model & scaler loaded")

FEATURES = [
    "WristAccelerometer_x-axis (g)",
    "WristAccelerometer_y-axis (g)",
    "WristAccelerometer_z-axis (g)",
    "WristAngularVelocity_x-axis (deg/s)",
    "WristAngularVelocity_y-axis (deg/s)",
    "WristAngularVelocity_z-axis (deg/s)",
    "Acc_mag",
    "Gyro_mag"
]

SEQ_LEN = 120

def prepare_input(filepath, subject, activity):
    df = pd.read_csv(filepath, header=[0,1])
    df.columns = ['_'.join(col) for col in df.columns]

    # Compute magnitude
    df["Acc_mag"] = np.sqrt(
        df["WristAccelerometer_x-axis (g)"]**2 +
        df["WristAccelerometer_y-axis (g)"]**2 +
        df["WristAccelerometer_z-axis (g)"]**2
    )

    df["Gyro_mag"] = np.sqrt(
        df["WristAngularVelocity_x-axis (deg/s)"]**2 +
        df["WristAngularVelocity_y-axis (deg/s)"]**2 +
        df["WristAngularVelocity_z-axis (deg/s)"]**2
    )

    activity_col = [c for c in df.columns if c.lower().startswith("activity")][0]
    subject_col  = [c for c in df.columns if "subject" in c.lower()][0]

    filt = df[(df[subject_col] == subject) & (df[activity_col] == activity)]

    if len(filt) < SEQ_LEN:
        print("Not enough data!")
        return None

    X = filt[FEATURES].values[:SEQ_LEN]
    X_scaled = scaler.transform(X)

    return X_scaled.reshape(1, SEQ_LEN, len(FEATURES))


subject = int(input("Enter Subject: "))
activity = int(input("Enter Activity: "))

X = prepare_input("DataSet.csv", subject, activity)

if X is not None:
    prob = MODEL.predict(X)[0][0]
    print("\nFall Probability:", prob)

    if prob >= 0.35:
        print("🚨 FALL DETECTED")
    else:
        print("✅ NOT FALL")
```
